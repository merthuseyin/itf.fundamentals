```python
def fizz_buzz(n):
  fblist = list(range(1, n+1))
  for i in range(n):
    if fblist[i] % 3 == 0 and fblist[i] % 5 == 0:
      fblist[i] = "FizzBuzz"
    elif fblist[i] % 3 == 0:
      fblist[i] = "Fizz"
    elif fblist[i] % 5 == 0:
      fblist[i] = "Buzz"
  print(* fblist, sep="\n")
fizz_buzz(100)
```
