```python
nmb = int(input("Enter a number to check: "))
if nmb != 1:
  for w in range(2, nmb):
    if nmb % w == 0:
      not_prime = True
      prime = False
      break
    else:
      not_prime = False
      prime = True
else:
  print(f"{nmb} is not a prime number")
print(not_prime * (f"{nmb} is not a prime number"))
print(prime * (f"{nmb} is a prime number."))
```
