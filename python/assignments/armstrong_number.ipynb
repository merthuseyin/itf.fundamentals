```python
num = input("Enter a number: ")
list_num = list(num)
n = len(list_num)
if num.isdigit():
  total = 0
  for i in list_num:
    total += int(i) ** n
  if total == int(num):
    print(f"{num} is an Armstrong number.")
  else:
    print(f"{num} is not an Armstrong number.")
else:
  print("It is an invalid entry. Don't use non-numeric, float, or negative values!")
```
